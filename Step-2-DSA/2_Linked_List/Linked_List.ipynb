```python
# Linked List
# Def : A linked list is a linear data structure where each element is a separate object.
# Each element (we will call it a node) of a list is comprising of two items - the data and a reference to the next node.
# The last node has a reference to null. The entry point into a linked list is called the head of the list.
# It should be noted that head is not a separate node, but the reference to the first node.
# If the list is empty then the head is null.

# Time Complexity
# Access : O(n)
# Search : O(n)
# Insertion at the beginning : O(1)
# Deletion at the beginning : O(1)
# Insertion at the end : O(n)
# Deletion at the end : O(n)
# Insertion in the middle : O(n)
# Deletion in the middle : O(n)
# Space Complexity : O(n)

# Types of Linked List
# 1. Singly Linked List
# 2. Doubly Linked List

# Linked List Operations
# 1. Insertion
# 2. Deletion
# 3. Search
# 4. Sort
# 5. Reverse
```
