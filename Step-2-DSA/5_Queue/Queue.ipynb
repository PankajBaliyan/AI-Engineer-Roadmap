```python
# Queue
# Queue is a linear data structure that follows the FIFO (First In First Out) principle.
# The elements are inserted at the rear end and are deleted from the front end of the queue.
# The queue is an abstract data type that is used to store the elements before processing them.
# The elements are processed in the same order in which they are inserted.
# The queue is used in the operating system for job scheduling and for handling the tasks.

# The queue is implemented using the following operations:
# Enqueue: It is used to insert an element into the queue.
# Dequeue: It is used to delete an element from the queue.
# IsEmpty: It is used to check whether the queue is empty or not.
# IsFull: It is used to check whether the queue is full or not.
# Peek: It is used to get the value of the front element without removing it.
# Size: It is used to get the number of elements in the queue.

# The queue can be implemented using the following methods:
# Array
# Linked List
# Circular Array
```

```python
from collections import deque

# Directly using the list

# q = deque()
# q.appendleft(1)
# q.appendleft(2)
# q.appendleft(3)
# print(q)
# q.pop()
# print(q)

# Using class
class Queue:
    def __init__(self):
        self.buffer = deque()

    def enqueue(self, val):
        self.buffer.appendleft(val)

    def dequeue(self):
        return self.buffer.pop()

    def is_empty(self):
        return len(self.buffer) == 0

    def size(self):
        return len(self.buffer)

q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.buffer)

q.dequeue()
print(q.buffer)

print(q.is_empty())
print(q.size())
```
