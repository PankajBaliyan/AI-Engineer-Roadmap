```python
arr = [64, 34, 25, 12, 22, 11, 90]

# Bubble sort: Sorting an array using bubble sort algorithm.
# It compares each pair of adjacent items and swaps them if they are in the wrong order, this process is repeated until no swaps are needed.

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# range() function is used to generate a sequence of numbers, here it generates a sequence of numbers from 0 to n-1.
# -1 for length, and -i for the number of elements that are already sorted.

print("Array before Sorting: ")
print(arr)
print("Array after Sorting: ")
print(bubble_sort(arr))
```

```python
arr = [64, 34, 25, 12, 22, 11, 90]

# SELECTION SORT: Sorting an array using selection sort algorithm.
# It divides the input array into two subarrays: a sorted subarray and an unsorted subarray.
# Then, it selects the smallest element from the unsorted subarray and swaps it with the leftmost unsorted element.

# Selection means assuming.
# In simple words: It assume the first element as the minimum and then compare it with the other elements in the array, if the element is smaller than the assumed minimum element, then it is swapped with the assumed minimum element.
# This process is repeated for each element in the array.

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print(selection_sort(arr))
```

```python
# INSERTION SORT:
arr = [64, 34, 25, 12, 22, 11, 90]

# Insertion Sort: Sorting an array using insertion sort algorithm, it builds the final sorted array one item at a time.
# It takes each element from the input array and places it in its correct position in the sorted array.

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

print(insertion_sort(arr))

# Algorithm - In simple words
# 1. Starts from the second element of the array.
# 2. i and key are on same index.
# 3. Compare the key with the elements before it (j) and if the key is smaller than the element, then shift the element to the right.
# 4. Repeat step 3 until the key is greater than the element.
# 5. Insert the key at the correct position.
# 6. Repeat the process for all elements in the array.
# 7. The array is sorted.
```
