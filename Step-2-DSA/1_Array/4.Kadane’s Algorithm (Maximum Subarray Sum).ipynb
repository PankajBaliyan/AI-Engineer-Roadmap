```python
# Kadaneâ€™s Algorithm (Maximum Subarray Sum)
# it, is a famous interview question, and it is a problem of dynamic programming.
# Given an array of integers, find the maximum sum of subsequence of the array.

# arr = [-2, -3, 4, -1, -2, 1, 5, -3]
arr = [1,3,5]

# Method 1:
def max_subarray_sum(arr):
    max_so_far = arr[0]
    max_ending_here = arr[0]
    for i in range(1, len(arr)):
        max_ending_here = max(arr[i], max_ending_here + arr[i])
        max_so_far = max(max_so_far, max_ending_here)
    return max_so_far

print(max_subarray_sum(arr))

# Method 2:
def kadane_algorithm(arr):
    n = len(arr)
    max_sum = 0
    for i in range(n):
        for j in range(i,n):
            sum = 0
            for k in range(i,j+1):
                sum += arr[k]
            if sum > max_sum:
                max_sum = sum
    return max_sum

print(kadane_algorithm(arr))
```
