```python
import matplotlib.pyplot as plt
```

```python
# plot() is a versatile command, and will take an arbitrary number of arguments. For example, to plot x versus y, you can issue the command:
# plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
# plt.plot([1, 12, 3, 4])

# The function title adds a title to the plot:
# plt.plot([1, 2, 13, 14], 'ro')

# plt.plot(x, y, label='Line Label', color='blue', linestyle='-', marker='o')
plt.plot([1, 2, 3, 4], [1, 12, 9, 16], label='Line Label', color='blue', linestyle='-', marker='o')
# label : is used to give a name to the line
# color : is used to give a color to the line
# linestyle : is used to give a style to the line
# marker : is used to give a marker to the line
plt.title('Line Plot')

# xlabel() and ylabel() are used to give a name to the x-axis and y-axis respectively.
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# legend() is used to show the legend of the plot
plt.legend()

plt.show()
```

```python
# scatter : is used to plot a scatter plot
# plt.scatter([1, 2, 3, 4], [1, 4, 9, 16], label="Scatter Plot", color="red", marker="o")

# bar : is used to plot a bar chart
# plt.bar([1, 2, 3, 4], [1, 4, 9, 16], label="Bar Chart", color="green")

# hist : is used to plot a histogram
# plt.hist([1, 2, 3, 4], [1, 4, 9, 16], label="Histogram", color="blue")

# pie : is used to plot a pie chart
plt.pie([1, 2, 3, 4], labels=["A", "B", "C", "D"], autopct="%1.1f%%")
# autopct : is used to show the percentage on the pie chart

plt.show()
```

```python
# figure() is used to create a new figure, and savefig() is used to save the figure
plt.figure()
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.savefig("plot.png")

plt.show()
```
