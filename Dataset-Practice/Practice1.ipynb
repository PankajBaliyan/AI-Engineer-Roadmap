	•	DecisionTreeRegressor is a machine learning algorithm from the sklearn library used for regression tasks.
	•	It creates a decision tree based on the training data and uses this tree to make predictions about the target variable.
	•	DecisionTreeRegressor(): This initializes the decision tree model for regression.
	•	random_state=1: This ensures that the model produces the same results every time you run the code, because decision trees can involve some randomness. Setting the random_state ensures reproducibility.
	•	The target variable y is the value you want the model to predict. In this case, you’re predicting SalePrice, which is presumably the price of houses.
	•	data_set1.SalePrice: This extracts the SalePrice column from the DataFrame data_set1 and assigns it to y.
	•	selected_columns: This is a list of column names you have chosen as the features (predictor variables) for your model.
	•	X = data_set1[selected_columns]: This selects the specified columns from the DataFrame data_set1 and assigns them to X. These features will be used to make predictions about the target variable y.
	
	•	.fit(X, y): This method trains (fits) the DecisionTreeRegressor model using the features (X) and the target variable (y).
	•	X: The input features (selected columns).
	•	y: The target variable (SalePrice).

	•	.predict(X): After the model has been trained, you can use it to make predictions. This method takes the input features (X) and returns the predicted target values.


```python
from sklearn.tree import DecisionTreeRegressor

# # Define model. Specify a number for random_state to ensure same results each run
iowa_model = DecisionTreeRegressor(random_state=1)

# y = subset.MSSubClass

# # selecting multiple columns from a dataframe
# selected_columns = ['MSZoning', 'LotFrontage', 'LotArea', 'Street']
# X = subset[selected_columns]


# # Fit model
# melbourne_model.fit(X, y)
y = data_set1.SalePrice

# selecting multiple columns from a dataframe
selected_columns = [
    "LotArea",
    "YearBuilt",
    "1stFlrSF",
    "2ndFlrSF",
    "FullBath",
    "BedroomAbvGr",
    "TotRmsAbvGrd",
]
X = data_set1[selected_columns]

# Fit model
iowa_model.fit(X, y)
# .fit() method to fit the model to the data, which is known as training the model.
# .fit() method requires two inputs: the features (X) and the target variable (y) which is the variable we're trying to predict.
# features (x) are the columns of data we use to make predictions.
# target variable (y) is the column of data we want to predict.

# Make a prediction
predictions = iowa_model.predict(X)
predictions
```

```python
# MODAL VALIDATION

# /1. SPLIT THE DATA INTO TRAINING AND VALIDATION DATA
# train_test_split

def split_data(X, y, test_size=0.2, random_state=42):
    """
    Split data into training and test sets.
    
    Args:
        X (numpy.ndarray): Features.
        y (numpy.ndarray): Target.
        test_size (float): Proportion of data to be used for testing.
        random_state (int): Random state for reproducibility.
        
    Returns:
        tuple: Tuple containing:
            X_train (numpy.ndarray): Training features.
            X_test (numpy.ndarray): Test features.
            y_train (numpy.ndarray): Training target.
            y_test (numpy.ndarray): Test target.
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# calling the function
X_train, X_test, y_train, y_test = split_data(X, y)

# /2. FIT THE MODEL TO THE TRAINING DATA
# Fit the model
iowa_model.fit(X_train, y_train)

# /3. MAKE PREDICTIONS ON THE VALIDATION DATA
# Make predictions
predictions = iowa_model.predict(X_test)
predictions

# print the top few actual prices from validation data
# print(y_test.head())

# /4. CALCULATE THE MEAN ABSOLUTE ERROR
# mean_absolute_error

def calculate_mae(y_true, y_pred):
    """
    Calculate mean absolute error.
    
    Args:
        y_true (numpy.ndarray): Actual target values.
        y_pred (numpy.ndarray): Predicted target values.
        
    Returns:
        float: Mean absolute error.
    """
    mae = mean_absolute_error(y_true, y_pred)
    return mae
mae = calculate_mae(y_test, predictions)
print(mae)
```

```python
# 2. With Model Validation (Using Train/Test Split)
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

# Load dataset
data_set1 = pd.read_csv(data_set1_path)  # Replace with your dataset

# Define target and features
y = data_set1["SalePrice"]
selected_columns = [
    "LotArea",
    "YearBuilt",
    "1stFlrSF",
    "2ndFlrSF",
    "FullBath",
    "BedroomAbvGr",
    "TotRmsAbvGrd",
]
X = data_set1[selected_columns]

# Split data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Define and train the model
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(X_train, y_train)

# Make predictions on the test set
test_predictions = iowa_model.predict(X_test)

# Calculate mean absolute error (MAE) on test set
mae_with_validation = mean_absolute_error(y_test, test_predictions)
print(f"Mean Absolute Error (With Validation): {mae_with_validation}")
```
